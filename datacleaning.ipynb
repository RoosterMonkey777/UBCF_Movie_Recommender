```python
import sys
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("data/movies_metadata.csv")
movies.head()
```

```python
#print(movies['id'].dtype)
#print(movies['imdb_id'].dtype)
```

```python
movies['id'] = movies['id'].astype(str).str.strip()
movies['imdb_id'] = movies['imdb_id'].astype(str).str.strip()
```

```python
#check whether there are any duplicate ids or imdb_ids
print(" duplicate id in total：", movies['id'].duplicated().sum())
print(" duplicate imbd_id in total：", movies['imdb_id'].duplicated().sum())
```

```python
# show all duplicated ids
dup_ids = movies[movies['id'].duplicated(keep=False)]['id'].unique()

# check if the content of duplicated id's other column is the same
for i in dup_ids:
    rows = movies[movies['id'] == i]
    if not rows.drop('id', axis=1).duplicated().all():
        print(f"id={i} need to be checked mannually:")
        print(rows)
```

```python
# found all duplicated ids, but they are exactly the same, so we can just drop duplicates
movies = movies.drop_duplicates(subset='id', keep='first')
```

```python
# find all duplicated imdb_ids
dup_imdb_ids = movies[movies['imdb_id'].duplicated(keep=False)]['imdb_id'].unique()
```

```python
# check if the content of duplicated imdb_id's other column is the same
for i in dup_imdb_ids:
    rows = movies[movies['imdb_id'] == i]
    if not rows.drop('imdb_id', axis=1).duplicated().all():
        print(f"imdb_id={i} need to be checked mannually:")
        print(rows)  
```

```python
# just like id, found all duplicated imdb_ids, but they are exactly the same, so we can just drop duplicates
movies = movies.drop_duplicates(subset='imdb_id', keep='first')
```

```python
# check again: id and imdb_id should not have any duplicates now (expected to be 0)
print("duplicate id ：", movies['id'].duplicated().sum())

print("duplicate imdb_id ：", movies['imdb_id'].duplicated().sum())
```

```python
# id column should not have any empty values or empty strings
print("empty id ：", movies['id'].isnull().sum())
print("id empty string：", (movies['id'].str.strip() == '').sum())
```

```python
# 2. check whether id is a pure digit
not_digit = ~movies['id'].str.isdigit()
print("in total: ", not_digit.sum(),"ids are not pure digit")
print("show them in details：")
print(movies.loc[not_digit, 'id'].head())
```

```python
#just drop all rows where id is not a pure digit
movies = movies[movies['id'].str.isdigit()]
```

```python
# drop rows where imdb_id is empty string or null
movies = movies[~movies['imdb_id'].isnull() & (movies['imdb_id'].str.strip() != '')]
```

```python
#imdb_id should be in the format of 'tt' followed by digits, e.g., 'tt1234567'
# check if imdb_id matches the expected format
invalid_imdb = ~movies['imdb_id'].fillna('').str.match(r'^tt\d+$')
print(invalid_imdb.sum(),"imdb_id can not match the expected format")
print("show them in details：")
print(movies.loc[invalid_imdb, 'imdb_id'].head())
```

```python
# check if imdb_id length is 9 characters (including 'tt' prefix)
invalid_length = ~movies['imdb_id'].str.match(r'^tt\d{7}$')
print( invalid_length.sum(),"imdb_ids' length is not 9 characters (including 'tt' prefix)")
print("show them in details：")
print(movies.loc[invalid_length, 'imdb_id'].head())
```

```python
movies = movies[movies['imdb_id'].str.match(r'^tt\d{7}$')]
```

```python
# check again if imdb_id length is 9 characters (expected to be 0)
invalid_length = ~movies['imdb_id'].str.match(r'^tt\d{7}$')
print( invalid_length.sum(),"imdb_ids' length is not 9 characters (including 'tt' prefix)")
```

```python
# check if id is a positive integer
print(movies[pd.to_numeric(movies['id'], errors='coerce') <= 0])
```

```python
# convert id to integer type, because ratings.csv movie_id is int
movies['id'] = movies['id'].astype(int)
```

```python
# handle another dataset: ratings.csv
ratings = pd.read_csv("data/ratings.csv")
print(ratings.head())
print(ratings.info())
```

```python
# if userId、movieId、rating missing, we will drop the row
missing_rows = ratings[['userId', 'movieId', 'rating']].isnull().any(axis=1).sum()
print(missing_rows,"missing rows will be deleted")
```

```python
# check how many rows's userId can not cast to int
invalid_userId = pd.to_numeric(ratings['userId'], errors='coerce').isnull()
print("found", invalid_userId.sum(),"rows userId can not cast to int")

# check how many rows's movieid can not cast to int
invalid_movieId = pd.to_numeric(ratings['movieId'], errors='coerce').isnull()
print("found ", invalid_movieId.sum(),"rows movieId can not cast to int")

# check how many rows's rating can not cast to float
invalid_rating = pd.to_numeric(ratings['rating'], errors='coerce').isnull()
print("found", invalid_rating.sum(),"rows rating can not cast to int")
```

```python
# cast userId, movieId, rating to int or float
ratings['userId'] = ratings['userId'].astype(int)
ratings['movieId'] = ratings['movieId'].astype(int)
ratings['rating'] = ratings['rating'].astype(float)
```

```python
# check whether there are any duplicate ratings (same userId and movieId)
print("duplicate rating pair：", ratings.duplicated(subset=['userId', 'movieId']).sum())
```

```python
#check if there are any ratings that are not in the range of 0 to 5
print(ratings[(ratings['rating'] < 0) | (ratings['rating'] > 5)])
```

```python
ratings = ratings.merge(movies[['id']], left_on='movieId', right_on='id', how='inner')
```

```python
# 找出 ratings 里 movieId 不在 movies['id'] 里的行
not_in_movies = ~ratings['movieId'].isin(movies['id'].astype(int))
print("ratings 里 movieId 在 movies_metadata 里找不到的数量：", not_in_movies.sum())
print("示例：")
print(ratings.loc[not_in_movies].head())
```
